# THEORY QUESTIONS - FUNCTIONS

**Q1. What is the difference between a function and a method in Python?**

**Example:**

- **Function**: A function is a block of code that is defined to perform a specific task. It can be called independently.

    ```python
    def add(a, b):
        return a + b

    result = add(5, 3)
    print(result)  # Output: 8
    ```

- **Method**: A method is a function that is associated with an object (usually an instance of a class). It operates on data within that object and is called using dot notation.

    ```python
    class Calculator:
        def add(self, a, b):
            return a + b

    calc = Calculator()
    result = calc.add(5, 3)  # Calling method on an object
    print(result)  # Output: 8
    ```



















**Q2. Explain the concept of function arguments and parameters in Python.**

**Q4. What is the purpose of the `return` statement in a Python function?**

The `return` statement is used to exit a function and return a result to the caller. Without `return`, the function will return `None` by default.

**Example:**

```python
def add(a, b):
    return a + b  # The function will return the result of a + b

result = add(5, 3)
print(result)  # Output: 8
```

**Q5. What are iterators in Python and how do they differ from iterables?**

- **Iterable**: An iterable is an object that can be looped over (like a list, tuple, or string). It implements the `__iter__()` method.

    **Example:**

    ```python
    my_list = [1, 2, 3]
    for item in my_list:  # my_list is an iterable
        print(item)
    ```

- **Iterator**: An iterator is an object that represents a stream of data and produces the next item when the `next()` function is called. It implements the `__next__()` method and the `__iter__()` method.

    **Example:**

    ```python
    my_iter = iter(my_list)  # Creating an iterator from a list
    print(next(my_iter))  # Output: 1
    print(next(my_iter))  # Output: 2
    ```

**Q6. Explain the concept of generators in Python and how they are defined.**

A generator is a special type of iterator in Python that yields items one at a time, allowing iteration without storing the entire sequence in memory. Generators are defined using `yield` instead of `return`.

**Example:**

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # yields the next value without exiting the function
        count += 1

counter = count_up_to(3)
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
```

**Q7. What are the advantages of using generators over regular functions?**

- **Memory Efficiency**: Generators do not require the entire sequence to be stored in memory, which is beneficial for large datasets.
- **Lazy Evaluation**: Items are computed only when requested, which can improve performance by avoiding unnecessary computation.
- **Convenience**: Generators can be used with `for` loops or directly with `next()`.


**Q8. What is a lambda function in Python and when is it typically used?**

A lambda function is an anonymous function defined using the `lambda` keyword. It can have any number of arguments but only one expression. It is often used for short, throwaway functions.

**Example:**

```python
multiply = lambda x, y: x * y
print(multiply(3, 4))  # Output: 12
```

**When used**: Lambda functions are often used as arguments to higher-order functions like `map()`, `filter()`, and `sorted()`.

**Q9. Explain the purpose and usage of the `map()` function in Python.**

The `map()` function applies a given function to all items in an iterable (e.g., list, tuple) and returns a map object (an iterator) of the results.

**Example:**

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared = map(square, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

**Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

- **`map()`**: Applies a function to each item of an iterable and returns a map object (an iterator) with the results.
  
    ```python
    map(function, iterable)
    ```

- **`reduce()`**: Applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, reducing it to a single value. It is found in the `functools` module.

    ```python
    from functools import reduce
    reduce(function, iterable)
    ```

- **`filter()`**: Filters the items in an iterable by applying a function that returns `True` or `False`, returning only those items for which the function returns `True`.

    ```python
    filter(function, iterable)
    ```

**Examples:**

- `map()`: Applies a function to each item.
  
    ```python
    map(lambda x: x * 2, [1, 2, 3])  # Output: [2, 4, 6]
    ```

- `reduce()`: Cumulatively reduces the list by applying a function.

    ```python
    from functools import reduce
    reduce(lambda x, y: x + y, [1, 2, 3, 4])  # Output: 10
    ```

- `filter()`: Filters items based on a condition.

    ```python
    filter(lambda x: x > 2, [1, 2, 3, 4])  # Output: [3, 4]
    ```

Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.



````python
numbers = input("Enter a list of number separated by spaces ")
numbers = numbers.split()
numbers = [int(num) for num in numbers]
def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even

print(sum_even_numbers(numbers))
````

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.



````python
specified_number=int(input("Enter the number of terms you want to generate the Fibonacci Sequence "))
def fibonacci_sequence (n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
        if a>specified_number:
          break

for num in fibonacci_sequence(specified_number):
    print(num)
````

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



````python
given_exp = int(input("Enter the exponent "))
def power_of_two(n):
    power = 1
    for _ in range(n):
        yield power
        power *= 2

for num in power_of_two(given_exp):
    print(num)
````

Q7. Implement a generator function that reads a file line by line and yields each line as a string.


````python
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip() #Use strip() to remove the newline character.
````

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.



````python
celsius_temperatures = [0, 5, 10, 15, 20, 25, 30]

#Function to convert Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

#Using map() to apply the conversion function to the list.
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
````

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.




````python
#Function to check if a character is a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

#Input string
input_string = "Hello, World!"

#Use filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
````

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
Write a Python program using lambda and map.


````python
# Sample data: List of orders where each order is a tuple (order_number, price_per_item, quantity)
orders = [
    (1, 40.95, 4),  # order number 1, price per item 40.95 €, quantity 4
    (2, 56.80, 5),  # order number 2, price per item 56.80 €, quantity 5
    (3, 32.95, 3),  # order number 3, price per item 32.95 €, quantity 3
    (4, 24.99, 3),  # order number 4, price per item 24.99 €, quantity 3
]

# Using map with a lambda function to calculate the order total and apply the additional 10 € if needed
order_totals = list(map(lambda order: (order[0], (order[1] * order[2]) + 10 if order[1] * order[2] < 100 else order[1] * order[2]), orders))

# Print the result
print(order_totals)
````
